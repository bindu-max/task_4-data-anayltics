```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset.xlsx")

# Display column names
print(df.columns)

# -------------------- 1. Bar Chart --------------------
plt.figure(figsize=(6,4))
df.groupby("Category")["Total_Sales"].sum().plot(kind="bar")
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()

# -------------------- 2. Pie Chart --------------------
plt.figure(figsize=(6,6))
df.groupby("Category")["Total_Sales"].sum().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.title("Total Sales Distribution by Category")
plt.ylabel("")
plt.show()

# -------------------- 3. Line Chart --------------------
plt.figure(figsize=(8,4))
df["Total_Sales"].plot(kind="line")
plt.title("Total Sales Trend")
plt.xlabel("Index")
plt.ylabel("Total Sales")
plt.show()



# -------------------- 5. Histogram --------------------
plt.figure(figsize=(6,4))
plt.hist(df["Total_Sales"], bins=15)
plt.title("Total Sales Distribution")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()

# -------------------- 6. Box Plot --------------------
plt.figure(figsize=(6,4))
plt.boxplot(df["Total_Sales"])
plt.title("Box Plot of Total Sales")
plt.ylabel("Total Sales")
plt.show()
```
